```python
# handle missing values
import numpy as np
import pandas as pd
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay']
}
df = pd.DataFrame(data)
print(df)
# null value filter | find out
print(df.isnull())

# sum of null
print(df.isnull().sum())

# dropna | drop complete row
df.dropna(inplace=True)
print(df)
```

```python
# handle missing values
import numpy as np
import pandas as pd
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay']
}
df = pd.DataFrame(data)
print(df)
# null value filter | find out
print(df.isnull())

# sum of null
print(df.isnull().sum())

# dropna | drop complete row
df.dropna()
print(df)

# fillna | null -> fill
df.fillna(5)
```

```python
# pie chart
import matplotlib.pyplot as plt #

x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]

plt.figure(figsize=(4,4))
plt.pie(y,labels=x,autopct="%1.1f%%",colors=["red","yellow","orange","green"],shadow=True)
plt.show()
```

```python
# subplots
import matplotlib.pyplot as plt #
# pie chart data
x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]
fig , aux = plt.subplots(1,2)
aux[0].pie(y,labels=x, colors=['red', 'yellow', 'orange', 'green'])

# line chart data
data = {
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500]
}
x = data['experience']
y = data['salary']

aux[1].plot(x,y)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ["Python", "Java", "C++", "C"]
marks = [85, 75, 90, 70]

fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# Line Chart
ax[0, 0].plot(subjects, marks, marker='o')
ax[0, 0].set_title("Line Chart")

# Bar Chart
ax[0, 1].bar(subjects, marks)
ax[0, 1].set_title("Bar Chart")

# Pie Chart
ax[1, 0].pie(marks, labels=subjects, autopct="%1.1f%%", startangle=180, shadow=True)
ax[1, 0].set_title("Pie Chart")

# Horizontal Bar Chart
ax[1, 1].barh(subjects, marks, color="skyblue")
ax[1, 1].set_title("Horizontal Bar Chart")

plt.tight_layout()
plt.show()
```
